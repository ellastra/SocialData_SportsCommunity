```python
import pandas as pd
import numpy as np
import os
import glob

print("=" * 70)
print("[ LCK ë°ì´í„° í†µí•© ]")
print("=" * 70)

# ë°ì´í„° ê²½ë¡œ ì„¤ì •
lck_data_path = r"C:\Users\qkrwl\Desktop\socialdata\SocialData_SportsCommunity\LCK_data"

# results_lck_1, 2, 3 CSV íŒŒì¼ ë¡œë“œ
results_files = [
    os.path.join(lck_data_path, "results_lck_1.csv"),
    os.path.join(lck_data_path, "results_lck_2.csv"),
    os.path.join(lck_data_path, "results_lck_3.csv")
]

# ê° íŒŒì¼ ë¡œë“œ
dfs = []
for file_path in results_files:
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        dfs.append(df)
        print(f"\nâœ“ {os.path.basename(file_path)} ë¡œë“œ ì™„ë£Œ")
        print(f"  í¬ê¸°: {df.shape[0]:,} rows Ã— {df.shape[1]} columns")
        print(f"  ì»¬ëŸ¼: {list(df.columns)}")
    else:
        print(f"\nâœ— {file_path} íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

# ë°ì´í„°í”„ë ˆì„ í•©ì¹˜ê¸°
if dfs:
    results_combined = pd.concat(dfs, ignore_index=True)
    
    print("\n" + "=" * 70)
    print("[ í†µí•© ê²°ê³¼ ]")
    print("=" * 70)
    print(f"\nâœ“ ë°ì´í„° í†µí•© ì™„ë£Œ")
    print(f"  ì´ í–‰ ìˆ˜: {results_combined.shape[0]:,}")
    print(f"  ì´ ì—´ ìˆ˜: {results_combined.shape[1]}")
    print(f"\nì»¬ëŸ¼ ì •ë³´:")
    for col in results_combined.columns:
        print(f"  - {col}")
    
    print(f"\në°ì´í„° ë¯¸ë¦¬ë³´ê¸°:")
    print(results_combined.head())
    
    print(f"\në°ì´í„° ì •ë³´:")
    print(results_combined.info())
else:
    print("ë¡œë“œëœ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

print("=" * 70)
print("[ ë‚ ì§œë³„ ê¸°ì´ˆí†µê³„ëŸ‰ ë¶„ì„ ]")
print("=" * 70)

# datetime_kstë¥¼ datetimeìœ¼ë¡œ ë³€í™˜
results_combined['datetime_kst'] = pd.to_datetime(results_combined['datetime_kst'])

# ë‚ ì§œë§Œ ì¶”ì¶œ
results_combined['date'] = results_combined['datetime_kst'].dt.date

# ë‚ ì§œë³„ ì§‘ê³„
daily_stats = results_combined.groupby('date').agg({
    'message': 'count',  # ë°ì´í„° ìˆ˜
    'polarity_binary': ['sum', 'count']  # ê¸ì •(1)ì˜ ê°œìˆ˜, ì „ì²´ ê°œìˆ˜
}).reset_index()

# ì»¬ëŸ¼ëª… ì •ë¦¬
daily_stats.columns = ['date', 'total_count', 'positive_count', 'total_count_check']
daily_stats['negative_count'] = daily_stats['total_count'] - daily_stats['positive_count']
daily_stats['positive_ratio'] = daily_stats['positive_count'] / daily_stats['total_count']
daily_stats['negative_ratio'] = daily_stats['negative_count'] / daily_stats['total_count']

print(f"\nâœ“ ë‚ ì§œë³„ ê¸°ì´ˆí†µê³„ëŸ‰ ê³„ì‚° ì™„ë£Œ")
print(f"  ë¶„ì„ ê¸°ê°„: {daily_stats['date'].min()} ~ {daily_stats['date'].max()}")
print(f"  ë¶„ì„ ì¼ìˆ˜: {len(daily_stats)}ì¼")

print(f"\nğŸ“Š ë‚ ì§œë³„ ê¸°ì´ˆí†µê³„ëŸ‰:")
print(daily_stats.to_string())

# í†µê³„ ìš”ì•½
print(f"\nã€ í†µê³„ ìš”ì•½ ã€‘")
print(f"  ì´ ëŒ“ê¸€ ìˆ˜: {daily_stats['total_count'].sum():,}")
print(f"  ê¸ì •(1) ëŒ“ê¸€: {daily_stats['positive_count'].sum():,} ({daily_stats['positive_count'].sum() / daily_stats['total_count'].sum() * 100:.1f}%)")
print(f"  ë¶€ì •(0) ëŒ“ê¸€: {daily_stats['negative_count'].sum():,} ({daily_stats['negative_count'].sum() / daily_stats['total_count'].sum() * 100:.1f}%)")
print(f"\n  ì¼ë³„ í‰ê·  ëŒ“ê¸€: {daily_stats['total_count'].mean():.0f}")
print(f"  ì¼ë³„ ìµœëŒ€ ëŒ“ê¸€: {daily_stats['total_count'].max():,} ({daily_stats.loc[daily_stats['total_count'].idxmax(), 'date']})")
print(f"  ì¼ë³„ ìµœì†Œ ëŒ“ê¸€: {daily_stats['total_count'].min():,} ({daily_stats.loc[daily_stats['total_count'].idxmin(), 'date']})")
```

```python
print("\n" + "=" * 70)
print("[ ë‚ ì§œë³„ ì‹œê°í™” - ê·¸ë˜í”„ 1: ì¼ì¼ ëŒ“ê¸€ ìˆ˜ ]")
print("=" * 70)

# daily_stats ì •ë ¬ ë° ë¦¬ì…‹
daily_stats_sorted = daily_stats.sort_values('date').reset_index(drop=True)

print(f"\nì •ë ¬ëœ ë‚ ì§œ ë²”ìœ„: {daily_stats_sorted['date'].min()} ~ {daily_stats_sorted['date'].max()}")
print(f"ì´ ì¼ìˆ˜: {len(daily_stats_sorted)}")

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style("whitegrid")

# ============================================================
# 1. ì¼ì¼ ëŒ“ê¸€ ìˆ˜ êº¾ì€ì„ ê·¸ë˜í”„ (ì²« ë²ˆì§¸ ì´ë¯¸ì§€)
# ============================================================
fig1, ax1 = plt.subplots(figsize=(16, 6))

x_indices = range(len(daily_stats_sorted))
ax1.plot(x_indices, daily_stats_sorted['total_count'].values, 
         marker='o', linewidth=2.5, markersize=6, color='#2E86AB', label='Total Comments')
ax1.fill_between(x_indices, daily_stats_sorted['total_count'].values, alpha=0.3, color='#2E86AB')
ax1.set_xlabel('Date', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of Comments', fontsize=12, fontweight='bold')
ax1.set_title('Daily Comment Count (Line Chart)', fontsize=14, fontweight='bold', pad=20)
ax1.grid(True, alpha=0.4, linestyle='--')

# xì¶• ë¼ë²¨ì„ ë‚ ì§œë¡œ í‘œì‹œ (ê°€ë…ì„±ì„ ìœ„í•´ ì¼ë¶€ë§Œ)
tick_positions = range(0, len(daily_stats_sorted), max(1, len(daily_stats_sorted)//15))
tick_labels = [str(daily_stats_sorted.loc[i, 'date']) for i in tick_positions]
ax1.set_xticks(tick_positions)
ax1.set_xticklabels(tick_labels, rotation=45, ha='right', fontsize=9)

# ìµœëŒ€/ìµœì†Œê°’ í‘œì‹œ
max_idx = daily_stats_sorted['total_count'].idxmax()
min_idx = daily_stats_sorted['total_count'].idxmin()
ax1.annotate(f"Max: {daily_stats_sorted.loc[max_idx, 'total_count']:,}\n({daily_stats_sorted.loc[max_idx, 'date']})", 
            xy=(max_idx, daily_stats_sorted.loc[max_idx, 'total_count']),
            xytext=(10, 20), textcoords='offset points',
            bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', color='red', lw=2),
            fontsize=10, fontweight='bold')
ax1.annotate(f"Min: {daily_stats_sorted.loc[min_idx, 'total_count']:,}\n({daily_stats_sorted.loc[min_idx, 'date']})", 
            xy=(min_idx, daily_stats_sorted.loc[min_idx, 'total_count']),
            xytext=(10, -40), textcoords='offset points',
            bbox=dict(boxstyle='round,pad=0.5', fc='lightblue', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', color='blue', lw=2),
            fontsize=10, fontweight='bold')

plt.tight_layout()

# ì €ì¥
output_dir = r"C:\Users\qkrwl\Desktop\socialdata\SocialData_SportsCommunity\LCK_PythonScript"
os.makedirs(output_dir, exist_ok=True)
output_file1 = os.path.join(output_dir, 'daily_comment_count.png')
plt.savefig(output_file1, dpi=300, bbox_inches='tight')
print(f"\nâœ“ ê·¸ë˜í”„ 1 ì €ì¥ ì™„ë£Œ: {output_file1}")

plt.show()
plt.close()
```
