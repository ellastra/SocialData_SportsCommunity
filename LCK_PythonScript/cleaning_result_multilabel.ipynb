```python
import pandas as pd
import os
# 상위 폴더에서 LCK_data 폴더 안의 emotion_predictions.csv 파일을 불러오기
file_dir = os.path.join('..', 'LCK_data', 'emotion_predictions.csv')

df = pd.read_csv(file_dir)
df.head()
```

```python
save_cols = ['datetime_kst','author','format','clean_text','emotion_probabilities','num_emotions']
df = df[save_cols].copy()

# emotion_probabilities 컬럼을 emotion_prob0~43 컬럼으로 분리
df['emotion_probabilities'] = df['emotion_probabilities'].str.strip('[]')
emotion_probs = df['emotion_probabilities'].str.split(',', expand=True)
emotion_probs.columns = [f'emotion_prob{i}' for i in range(emotion_probs.shape[1])]
df = pd.concat([df.drop(columns=['emotion_probabilities']), emotion_probs], axis=1)
df.sample(5)
```

```python
labels = ['불평/불만', '환영/호의', '감동/감탄', '지긋지긋', '고마움', '슬픔', '화남/분노', '존경', '기대감', '우쭐댐/무시함',
          '안타까움/실망', '비장함', '의심/불신', '뿌듯함', '편안/쾌적', '신기함/관심', '아껴주는', '부끄러움', '공포/무서움',
          '절망', '한심함', '역겨움/징그러움', '짜증', '어이없음', '없음', '패배/자기혐오', '귀찮음', '힘듦/지침', '즐거움/신남',
          '깨달음', '죄책감', '증오/혐오', '흐뭇함(귀여움/예쁨)', '당황/난처', '경악', '부담/안_내킴', '서러움', '재미없음',
          '불쌍함/연민', '놀람', '행복', '불안/걱정', '기쁨', '안심/신뢰']

# emotion_prob0~43 컬럼에 라벨명 지정
for i, label in enumerate(labels):
    df.rename(columns={f'emotion_prob{i}': label}, inplace=True)
df.sample(5)
```

```python
df.to_csv(os.path.join('..', 'LCK_data', 'emotions_probs.csv'), index=False)
```

```python
# 임계값 이상의 감정만 one-hot encoding하는 함수 작성하기
def one_hot_encode_emotions(df, threshold=0.5):
    emotion_cols = df.columns.difference(['datetime_kst', 'author', 'format', 'clean_text', 'num_emotions'])
    for col in emotion_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = (df[col] >= threshold).astype(int)
    return df

df = one_hot_encode_emotions(df, threshold=0.5)
```

```python
df.sample(10)
```

```python
df.to_csv(os.path.join('..', 'LCK_data', 'emotions_one_hot.csv'), index=False)
```
